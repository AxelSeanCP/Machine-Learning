### Handle outliers with IQR method
- bottom threshold = Q1 - 1.5 * IQR
- top threshold = Q3 + 1.5 * IQR

```python
Q1 = diamonds.quantile(0.25, numeric_only=True)
Q3 = diamonds.quantile(0.75, numeric_only=True)

IQR = Q3 - Q1
left_Q1, right_Q1 = diamonds.align((Q1 - 1.5 * IQR), axis=1, copy=False)
left_Q3, right_Q3 = diamonds.align((Q3 + 1.5 * IQR), axis=1, copy=False)
diamonds = diamonds[~((left_Q1 < right_Q1) | (left_Q3 > right_Q3)).any(axis=1)]

# Check dataset size after outliers were dropped
diamonds.shape
```
